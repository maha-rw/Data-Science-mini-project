```python
#importing required Libraries
import pandas as pd   
import requests       
from bs4 import BeautifulSoup 
import numpy as np  
import matplotlib.pyplot as plt 
```

COLLECTING using web-scraping

```python
url = 'https://www.imdb.com/search/title/?count=250&groups=top_1000&sort=user_rating'
url2 = 'https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc&count=250&start=251&ref_=adv_nxt'
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
response2 = requests.get(url) #for the second page
soup2 = BeautifulSoup(response2.content, 'html.parser')
```

```python
print(soup)
orint("\n\n\n\n")
print(soup2)
```

```python
#lists so that we can append the values
movie_name = []
year = []
duration = []
rating = []
votes = []
gross = []
Director = []
Stars = []
```

```python
movie_data = soup.findAll('div', attrs= {'class': 'lister-item mode-advanced'})#this class is the class of the <div> element that represent a single movie(single row)
movie_data2 = soup2.findAll('div', attrs= {'class': 'lister-item mode-advanced'})#this class is the class of the <div> element that represent a single movie(single row)
```

```python
 #CLEANING and COLLECTING at the same time by changing columns to appropriate types and forms.
for store in movie_data: #calling row by row using for loop

    name = store.h3.a.text # name of the movie stored inside <h3> <a> tags
    movie_name.append(name)
    
    
    year_of_release_str = store.h3.find('span', class_ = 'lister-item-year text-muted unbold').text.replace('(', '').replace(')', '').replace('II','').replace('I','')
    year_of_release = int(year_of_release_str)
    year.append(year_of_release)
    
    
    runtime_str = store.p.find('span', class_ = 'runtime').text.replace(' min', '')
    runtime = int( runtime_str)
    duration.append(runtime)
    
    rate_str = store.find('div', class_ = 'inline-block ratings-imdb-rating').text.replace('\n', '')
    rate=float(rate_str)
    rating.append(rate)
    
    # gross and votes have same attributes, that's why we had created a common variable and then used indexing
    value = store.find_all('span', attrs = {'name': 'nv'})
    
    vote_str = value[0].text.replace(',','')# to make it able to be casted to integer
    vote= int(vote_str)
    votes.append(vote)
    
    grosses_str = value[1].text.replace('M','').replace('$','') if len(value)>1 and "#" not in value[1].text else 'NaN'
    grosses=float(grosses_str)
    gross.append(grosses)
   
    #Cast Details -- Scraping Director name and Stars -- Not explained in Video
    cast = store.find("p", class_ = '')
    cast = cast.text.replace('\n', '').split('|')
    cast = [x.strip() for x in cast]
    cast = [cast[i].replace(j, "") for i,j in enumerate(["Director:", "Stars:"])]
    Director.append(cast[0])
    Stars.append([x.strip() for x in cast[1].split(",")])
    
for store in movie_data2: #calling row by row using for loop for the 2nd page
    name = store.h3.a.text # name of the movie stored inside <h3> <a> tags
    movie_name.append(name)
    
    
    year_of_release_str = store.h3.find('span', class_ = 'lister-item-year text-muted unbold').text.replace('(', '').replace(')', '').replace('II','').replace('I','')
    year_of_release = int(year_of_release_str)
    year.append(year_of_release)
    
    
    runtime_str = store.p.find('span', class_ = 'runtime').text.replace(' min', '')
    runtime = int( runtime_str)
    duration.append(runtime)
    
    rate_str = store.find('div', class_ = 'inline-block ratings-imdb-rating').text.replace('\n', '')
    rate=float(rate_str)
    rating.append(rate)
    
    # gross and votes have same attributes, that's why we had created a common variable and then used indexing
    value = store.find_all('span', attrs = {'name': 'nv'})
    
    vote_str = value[0].text.replace(',','')# to make it able to be casted to integer
    vote= int(vote_str)
    votes.append(vote)
    
    grosses_str = value[1].text.replace('M','').replace('$','') if len(value)>1 and "#" not in value[1].text else 'NaN'
    grosses=float(grosses_str)
    gross.append(grosses)
   
    #Cast Details -- Scraping Director name and Stars -- Not explained in Video
    cast = store.find("p", class_ = '')
    cast = cast.text.replace('\n', '').split('|')
    cast = [x.strip() for x in cast]
    cast = [cast[i].replace(j, "") for i,j in enumerate(["Director:", "Stars:"])]
    Director.append(cast[0])
    Stars.append([x.strip() for x in cast[1].split(",")])
    
    
print(name)
print("\n\n")

print(year)
print("\n\n")

print(duration)
print("\n\n")

print(rating)
print("\n\n")

print(votes)
print("\n\n")

print(gross)
print("\n\n")

print(Director)
print("\n\n")

print(Stars)
    
    
```

```python
#creating a dataframe using pandas library
movie_DF = pd.DataFrame({'Movie Name': movie_name, 'Year of release': year, 'Watchtime': duration, 'Movie Rating': rating, 'Votes': votes, 'Gross collection': gross,  'Director': Director, 'Cast': Stars})
```
